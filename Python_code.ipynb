```python
import pandas as pd
import psycopg2
from psycopg2.extras import execute_values

## Database configuration
DB_CONFIG = {
    "host":"localhost",
    "database":"Sales_DB",
    "user":"postgres",
    "password":"Postgr@0909",
    "port":5432
}
### Extract Read_csv:
df = pd.read_csv("Product_Data_set1.csv")
print("Product_data:")
print(df)

## Load - Insert into postgreSQL
conn = psycopg2.connect(**DB_CONFIG)
cur = conn.cursor()

## Insert Data (Efficient Way)'
records = df.to_records(index=False).tolist()

## Insert Data (Efficient way)
insert_query = """
INSERT INTO product_db(orderdate,order_number,quantity_ordered,price_each,product_name,product_line,buy_price
,city,country,sales_values,cost_of_sales,net_profit,total_revenue) VALUES %s
ON CONFLICT (order_number) DO NOTHING;
"""
execute_values(cur, insert_query,records)
conn.commit()
cur.close()
conn.close()

print("\n Data successfully loaded into postgresSQL!")
```
